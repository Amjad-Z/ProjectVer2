```python
s =requests.Session()
s.headers.update({"User-Agent": random.choice(UA)}) 
s.headers.update({'referer': referer})
s.headers.update({'host': host})
s.headers.update({'origin': origin})
s.headers
```

### Scraping the first page of players:

```python
api = 'https://api.sofascore.com/api/v1/unique-tournament/7/season/29267/statistics?limit=20&order=-rating&group=summary'
```

```python
pages = None
data_list = []
r = s.get(api)
if r.status_code==200:
    data = json.loads(r.text)
    data_list.append(data)
    pages = data['pages']
    
print(pages)
```

```python
data_list
```

### Scraping all the pages:

```python
offset = 0

for i in range(0, pages):
    offset += 20
    api = f'https://api.sofascore.com/api/v1/unique-tournament/7/season/29267/statistics?limit=20&order=-rating&offset={offset}&group=summary'
    s.headers.update({"User-Agent": random.choice(UA)}) 
    r = s.get(api)
    print(api, r.status_code)
    d = json.loads(r.text)
    data_list.append(d)
    time.sleep(5)
```

### Testing the output in a dictionary:

```python
players_stats = {}
for i in range(37):
 for player in data_list[i]['results']:
    p_id = player['player']['id']
    players_stats['id'] = p_id
    players_stats[p_id] = {}
    players_stats[p_id]['name'] = player['player']['name']
    players_stats[p_id]['club'] =player['team']['slug']
    players_stats[p_id]['rating'] =player['rating']
```
